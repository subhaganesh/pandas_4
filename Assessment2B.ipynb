###### Department wise codekata performence (pie chart)

```python
df.columns
df['Department']
df.groupby(['Department']).count()
```

```python
cse = df.loc[df['Department'] == 'Computer Science and Engineering'].count()['CodeKata Score']
ece = df.loc[df['Department'] == 'Electronics and Communication Engineering'].count()['CodeKata Score']
eee=df.loc[df['Department'] == 'Electronics and Electrical Engineering'].count()['CodeKata Score']

plt.figure(figsize=(8,5),dpi=100)

labels = ['cse', 'ece','eee']
colors = ['#abcdef', '#aabbcc','c']

plt.pie([cse, ece,eee], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('codekata performance',fontdict={'fontname': 'Comic Sans MS', 'fontsize': 30})

plt.show()
```
